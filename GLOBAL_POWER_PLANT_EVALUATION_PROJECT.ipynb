```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
def vif_check (df,target):
    x=df.drop(target,axis=1)
    vif=pd.DataFrame()
    vif['vif']=[variance_inflation_factor(x,i) for i in range(x.shape[1])]
    vif['features']=x.columns
    print(vif)
```

```python
def algo_class_check (x,y,algo):
    max_acc=0
    max_i=0
    for i in range(100):
        x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state = i)
        algo.fit(x_train,y_train)
        y_pred = algo.predict(x_test)
        acc= accuracy_score(y_test,y_pred)
        if acc> max_acc:
            max_acc=acc
            max_i = i
            i+=1
    y=label_binarize(y,classes=class1)

    x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state = max_i)
    algo1=OneVsRestClassifier(algo)
    algo1.fit(x_train,y_train)
    y_pred = algo1.predict(x_test)
    acc= accuracy_score(y_test,y_pred)
    cvs=cross_val_score(algo1,x_train,y_train,cv=7,scoring='accuracy')
    ac=cvs.mean()
    roc_auc={}
    fpr=dict()
    tpr=dict()
    y_score=algo1.predict_proba(x_test)
    plt.figure(figsize=(15,15),facecolor='yellow')
    plotno=1
    for i in range(len(class1)):
        fpr[i],tpr[i],_=roc_curve(y_test[:,i],y_score[:,i])
        roc_auc[i]=auc(fpr[i],tpr[i])
        plt.subplot(3,3,plotno)
        plt.plot(fpr[i],tpr[i],color='green',label='ROC Curve (area = %0.2f)' %roc_auc[i])
        plt.plot([0,1],[0,1],linestyle='--',color='orange',lw=2)
        plt.xlim([-0.05,1.05])
        plt.ylim([0.0,1.05])
        plt.xlabel('False +ve Rate')
        plt.ylabel('True +ve Rate')
        plt.title(f"Roc curve ovr for class {i}")
        plt.legend(loc='lower right')
        plotno+=1
    plt.show()
    roc_auc_ovr=0
    for i in range(len(class1)):
        roc_auc_ovr+=roc_auc[i]
        #roc_auc_score(y_test,y_proba,multi_class='ovr',average='macro')
    #wighted_roc_auc_ovr=roc_auc_score(y_test,y_proba,multi_class='ovr',average='weighted')
    print(f'''for algo {algo}, the maximum accuracy is {max_acc} ,\n classification report is : \n {classification_report(y_test,y_pred)} ,
    \n at random_state{max_i} \n and cross validation score is {ac} \none vs rest average roc_auc_score is {roc_auc_ovr/8} 
    ''') 
```

```python
def algo_check (x,y,algo):
    min_diff=1
    max_i=0
    for i in range(100):
        x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state = i)
        algo.fit(x_train,y_train)
        y_pred1 =algo.predict(x_train)
        acc1= r2_score(y_train,y_pred1)
        y_pred2 =algo.predict(x_test)
        acc2= r2_score(y_test,y_pred2)
        acc=acc1-acc2
        if acc< min_diff:
            min_diff=acc
            max_i = i
            i+=1
    x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state = max_i)
    algo.fit(x_train,y_train)
    y_pred1 =algo.predict(x_train)
    acc1= r2_score(y_train,y_pred1)
    y_pred2 =algo.predict(x_test)
    acc2= r2_score(y_test,y_pred2)
    cvs=cross_val_score(algo,x_train,y_train,cv=5,scoring='r2')
    ac=cvs.mean()
    mae=mean_absolute_error(y_pred2,y_test)
    mse=mean_squared_error(y_pred2,y_test)
    rmse=np.sqrt(mse)
    print(f'''for algo {algo}, \nthe training accuracy is {acc1}, \ntesing accuracy is {acc2} \nat random_state {max_i} 
        \nand hence, mean square error is {mse} \nand mean_absolute_error is {mae} \nand hence, rmse is {rmse}, \nalso cross_validation_score is {ac}''')
```
